```python
import pandas as pd
```

```python
df = pd.read_csv('./ten_year.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
# convert datetime to only date
df['update'] = pd.to_datetime(df['update'], format='%d/%m/%Y %H:%M:%S').dt.strftime('%Y-%m-%d')
df.head()
```

```python
# SELECT MAX(sell) FROM df GROUP BY (update,type)
tmp = df.groupby(['update','type']).max().reset_index()
tmp
```

```python
tmp.to_csv('ten_year_max.csv', index=False)
```

```python
df2 = pd.read_csv('./six_month.csv')

# convert datetime to only date
df2['update'] = pd.to_datetime(df2['update'], format='%d/%m/%Y %H:%M:%S').dt.strftime('%Y-%m-%d')
df2.head()

# SELECT MAX(sell) FROM df2 GROUP BY (update,type)
sixmonthmax = df2.groupby(['update','type']).max().reset_index()
sixmonthmax

sixmonthmax.to_csv('six_month_max.csv', index=False)
```

```python
sixmonthmaxPNJ = sixmonthmax[sixmonthmax['type'] == 'PNJ']
sixmonthmaxSJC = sixmonthmax[sixmonthmax['type'] == 'SJC']

sixmonthmaxPNJ.to_csv('six_month_max_PNJ.csv', index=False)
sixmonthmaxSJC.to_csv('six_month_max_SJC.csv', index=False)
```
